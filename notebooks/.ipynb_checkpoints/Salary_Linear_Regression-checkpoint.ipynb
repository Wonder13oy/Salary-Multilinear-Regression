# Hypothesis Testing
**Null theory** - States that there is no correlation between the salary and the other columns

**Alternate theory** - States that there is atleast one correlation between salary and the other columns

```python
sns.pairplot(df);
```

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

```python
df = df[['yearsworked', 'yearsrank', 'position', 'salary']]
```

# P value and Variance

```python
print(f'p-values between years-worked and salary: {stats.pearsonr(df.yearsworked, df.salary)}')
print(f'p-values between years-rank and salary: {stats.pearsonr(df.yearsrank, df.salary)}')
print(f'p-values between postion and salary: {stats.pearsonr(df.position, df.salary)}')
```

We have found that there are variables that are highly correlated to `salary`. We found that variables are `yearsworked`, `yearsrank` and `position`. The correlation and p-values are shown above, as you can see all the p-values are below `0.05`. We therefore reject the **Null Hypothesis**.

# IV vs IV

```python
print(f'p-values between years-worked and years-rank: {stats.pearsonr(df.yearsworked, df.yearsrank)}')
print(f'p-values between years-rank and position: {stats.pearsonr(df.yearsrank, df.position)}')
print(f'p-values between postion and years-worked: {stats.pearsonr(df.position, df.yearsworked)}')
```

```python
sns.pairplot(df);
```

The above graphs show us that there is **Multicolinearity**. This means that there is a high correlation between the independent variables. This poses a problem of we do not know whether the Independent variable influences the dependent variable or does it influence the other variables. The biggest cases found are:

- `yearsworked` and `yearsrank`
- `yearsworked` and `position`

This will now pose the question of should I drop `yearsworked` or should I remove the other `variables`?

# Modeling

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import sklearn.model_selection
```

```python
X = df[['yearsworked','yearsrank','position']]
y = df['salary']

X = sm.add_constant(X)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)

model = sm.OLS(y_train, x_train).fit()
```

I have split the dataset in the ratio of 7:3(training to testing). It is common standard that 30% of your data should be able to explain 70% of your dataset.

# Dealing with Catogorical Data

```python
def replaceNum(num):
    if num == 1:
        return 'Junior'
    elif num == 2:
        return 'Manager'
    else:
        return 'Executive'
```

```python
df_modified = df['position'].apply(replaceNum)
```

```python
df['position'] = df_modified
df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

I have used the `pd.get_dummies()` to seperate the catagorical data. Each catagory will now be a feature and will be denoted with either 0 or 1 (no or yes). This is to show whether the individual is either a *junior employee*, *manager* or an *executive*.

```python
exec_df = df[df['position_Executive'] == 1]
jr_df = df[df['position_Junior'] == 1]
man_df = df[df['position_Manager'] == 1]
```

```python
plt.scatter(exec_df['yearsworked'], exec_df['salary'], marker='^');
plt.scatter(jr_df['yearsworked'], jr_df['salary'], marker='x');
plt.scatter(man_df['yearsworked'], man_df['salary'], marker='o');

plt.xlabel('years worked')
plt.ylabel('salary')
plt.legend(['Executive', 'Junior', 'Manager'],loc='best')
plt.show()
```

From the scatterplot above, you can see the relationship each position has with `salary`. You can see that this scatterplot looks way better than the first one we had. You can see that there *junior employees* are more concentrated near the bottom. This means that more *junior employees* have worked for less years and earn less. You also can see that *executives* have worked for more years and earn more. This shows that there is a relationship between `yearsworked` and `position`

```python
X = df.drop(columns=['salary'])
y = df['salary']

X = sm.add_constant(X)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)

all_model = sm.OLS(y_train, x_train).fit()
all_model.summary()
```

I have ran the model with all the independent variables. This is done to show much of an effect **multicollinearity** has on the model outcomes.

```python
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
```

```python
vif
```

The Variance Inflation Factor (VIF) is a measure of colinearity among predictor variables within a multiple regression. It is calculated by taking the the ratio of the variance of all a given model's betas divide by the variane of a single beta if it were fit alone. If there *VIF* is of features is under 5, then there is little to no **multicollinearity**. If it ranges from 5 to 10, then there is medium **multicollinearity**. Anything above 10 shows that there is a **HIGH** case of **multicollinearity**.
